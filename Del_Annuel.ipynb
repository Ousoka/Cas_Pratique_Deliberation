```python
import os
import pandas as pd
import numpy as np
cwd = os.getcwd()
print(cwd)
os.chdir("C:\\Users\\LENOVO\\Desktop\\Informatique\\12-05-2022")
fichier = "Del_Annuel.xlsx"
xl = pd.ExcelFile(fichier)
deliberation=xl.parse("PV")
deliberation
```

```python
# print(deliberation)
```

```python
idx=deliberation.index[deliberation["Université Amadou Mahtar MBOW"] == 'Nom'].tolist()[0]
idx
```

```python
title=deliberation.iloc[idx,:]
title
```

```python
deliberation1=deliberation.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
deliberation1
```

```python
deliberation2=deliberation1.rename(columns=title)
deliberation2
```

```python
deliberation2.columns=deliberation2.columns.fillna("Decision")
deliberation2.rename(columns = {'RESULTAT':'TotalCredits'}, inplace = True)
deliberation2
```

```python
deliberation2.reset_index(drop=True)
```

```python
tri_deliberation2=deliberation2.sort_values(by=['TotalCredits','Nom','Prénom'], ascending=[False,True,True])
tri_deliberation2
```

```python
tri_deliberation2=tri_deliberation2.reset_index(drop=True)
tri_deliberation2
```

```python
Eff=tri_deliberation2.Decision.value_counts()
Eff=pd.DataFrame(Eff)
Eff.rename(columns = {"Decision":"Nombre_Etudiants"}, inplace = True)
Pourcentages= (Eff['Nombre_Etudiants'] /Eff['Nombre_Etudiants'].sum()) * 100
Eff
```

```python
import math
def round_up(n, decimals=1):
    multiplier = 10 ** decimals
    return math.ceil(n * multiplier) / multiplier
def truncate(n, decimals=1):
    multiplier = 10**decimals
    return int(n* multiplier)/multiplier
P=[]
for i in range(len(Pourcentages)):
    if i==1:
        x=round_up(Pourcentages[i])
        P.append(x)
    else:
        x=truncate(Pourcentages[i])
        P.append(x)
print(P)
# truncate(24.444, decimals=1)
```

```python
Eff['Pourcentages'] = P
Eff.loc["Total"]=[sum(Eff.Nombre_Etudiants),sum(Eff.Pourcentages)]
Eff
```

```python
writer=pd.ExcelWriter("Déliberation.xlsx",engine="xlsxwriter")
tri_deliberation2.to_excel(writer,"Résultats_Globaux")
Eff.to_excel(writer,"Statistiques")
writer.save()
```

```python
# Import libraries
from matplotlib import pyplot as plt
import numpy as np
Labels = ['Passage Définitif', 'Redouble', 'Passage Conditionnel','Exclu']
data = P
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
plt.pie(data, labels = Labels)
# show plot
plt.show()
```

```python
#Second_Method
from matplotlib import pyplot as plt
import numpy as np
colors=('gold','cyan','pink','grey')
wp={'linewidth':1,'edgecolor':'green'}
fig,ax=plt.subplots(figsize=(10,7))
wedges,texts,autotexts=ax.pie(Eff.Pourcentages[:4],labels=Eff[:4].index,explode=(0.05,0.05,0.05,0.05),autopct='%1.1f%%',startangle=90,shadow=True,wedgeprops=wp,textprops=dict(color='darkblue'),colors=colors)
ax.legend(wedges,Eff[:4].index,title='Résultats Globaux',loc='center left',bbox_to_anchor=(1,0,0.5,1))
plt.setp(autotexts,size=8,weight='bold')
ax.set_title('Diagramme circulaire des Résultats Globaux')
plt.axis('equal')
plt.show()
```
